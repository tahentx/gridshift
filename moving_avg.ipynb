```python
# import various packages
import os
import scipy
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    plt.plot(a, label="Raw Values")
    plt.plot(ret[n - 1:] / n, label="Moving Avg")
    plt.xlabel("Time")
    plt.ylabel("Actual vs. Moving Averge")
    plt.title("Data Science Interview Q#2")
    plt.legend()
    plt.show()
#     return min(ret[n - 1:] / n),max(ret[n - 1:] / n) 
```

```python
t =  [4, 4, 4, 9, 10, 11, 12]
moving_average(t)
```
